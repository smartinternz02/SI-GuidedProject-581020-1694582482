```python
#scatter plot
sns.scatterplot(x="alcohol",y="total",data=df)
```

```python
sns.histplot(data=df, x='speeding', kde=True)
```

Inference:

 This histogram displays the distribution of speeding-related car crashes. The KDE (Kernel Density Estimation) curve suggests a peak in the middle range of speeding incidents.

```python
#scatter plot
sns.scatterplot(x="total",y="not_distracted",data=df)
```

Inference:

The scatter plot between "total" car crashes and "not_distracted" drivers in the dataset shows a scattered distribution with no strong linear relationship, suggesting that distraction levels may not be a significant predictor of total car crashes.

```python
#line plot
sns.lineplot(y="alcohol",x="total",data=df,ci=None)
```


Inference:

The line plot depicting the relationship between the "total" number of car crashes and "alcohol" consumption in the dataset, with confidence intervals disabled (ci=None), shows a general trend that suggests an increase in alcohol consumption corresponds to a potential increase in the total number of car crashes

```python
#distribution plot (histogram)
sns.distplot(df["total"])
```

Inference:

The sns.distplot() for the "total" variable in the dataset displays a histogram along with a Kernel Density Estimate (KDE) curve,

```python
#relational plot (scatter plot)
sns.relplot(x="total",y="alcohol",data=df.head(15),hue="abbrev")
```


Inference:

The sns.relplot() visualizes the relationship between the "total" number of car crashes and "alcohol" consumption in the dataset while differentiating data points by "abbrev" using different colors, allowing for the examination of potential regional variations in this relationship.

```python
#correlation
corr=df.corr()
corr
```

Inference:

The "corr" DataFrame contains correlation coefficients that quantify the linear relationships between pairs of variables in the original DataFrame "df."
##>0.5 is highly correlated
##<0.5 is less correlated

```python
#heatmap
sns.heatmap(corr,annot=True,cmap="YlGnBu")
```

Inference:

This code creates a heatmap visualization of the correlation matrix "corr," with numerical annotations, using the "YlGnBu" color map to represent the strength and direction of relationships between variables.



```python
#joint plot (combines scatter plot and historgram)
sns.jointplot(x="not_distracted",y="alcohol",data=df)
```

Inference:

This code generates a joint plot to explore the relationship and distributions between "not_distracted" and "alcohol" variables in the DataFrame "df."

```python
#pie chart
attributes = ['Total', 'Speeding', 'Alcohol', 'Not Distracted', 'No Previous']
attribute_values = [df['total'].sum(), df['speeding'].sum(), df['alcohol'].sum(), df['not_distracted'].sum(),
                    df['no_previous'].sum()]
plt.figure(figsize=(5, 5))
plt.pie(attribute_values, labels=attributes, autopct='%1.1f%%', startangle=140, colors=["#FF5733", "#42A5F5", "#9C27B0", "#FF9800", "#795548"])
plt.axis('equal')
```

```python
#explode pie chart

Explode=[0.3,0.3,0.3,0.3,0.3]

attributes = ['Total', 'Speeding', 'Alcohol', 'Not Distracted', 'No Previous']
attribute_values = [df['total'].sum(), df['speeding'].sum(), df['alcohol'].sum(), df['not_distracted'].sum(),
                    df['no_previous'].sum()]
plt.figure(figsize=(5, 5))
plt.pie(attribute_values, labels=attributes,explode=Explode, autopct='%1.1f%%', startangle=140, colors=["#FF5733", "#42A5F5", "#9C27B0", "#FF9800", "#795548"])
plt.axis('equal')
```

Inference:

This code generates a pie chart that displays the proportions of various attributes in the dataset, with labels and percentages, using specified colors.
